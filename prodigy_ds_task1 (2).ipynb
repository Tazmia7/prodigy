#Distribution of the continuous variable

#Distribution of country population

```python
plt.figure(figsize=(10,6))
plt.hist(data['2022 Population'],bins=20,color='orange',edgecolor='black')
plt.title('Distribution of Country Population(2022)')
plt.xlabel('Population')
plt.ylabel('Number of countries')
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.hist(data['2020 Population'],bins=20,color='skyblue',edgecolor='black')
plt.title('Distribution of Country Population(2020)')
plt.xlabel('Population')
plt.ylabel('Number of countries')
plt.show()
```

#Distribution of the categorical variable


#Total population of 2022 by continent

```python
continent_pop22 = data.groupby('Continent')['2022 Population'].sum().sort_values(ascending=False)
continent_pop22
```

```python
plt.figure(figsize=(10,6))
continent_pop22.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('Total Population by Continent(2022)')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=continent_pop22.index,y=continent_pop22.values,palette='coolwarm')
plt.title('Total Population by Continent(2022)')
plt.xlabel('Continent')
plt.ylabel('Total Population(2022)')
plt.show()
```

#Total population of 2020 by Continent

```python
continent_pop20 = data.groupby('Continent')['2020 Population'].sum().sort_values(ascending=False)
continent_pop20
```

```python
plt.figure(figsize=(10,6))
continent_pop20.plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('Total Population by Continent(2020)')
plt.xlabel('Continent')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=continent_pop20.index,y=continent_pop20.values,palette='coolwarm')
plt.title('Total Population by Continent(2020)')
plt.xlabel('Continent')
plt.ylabel('Total Population(2020)')
plt.show()
```

#Top-10 countries with respect to total population

```python
total_pop=data.groupby('Country/Territory')['2022 Population'].sum().sort_values(ascending=False)
total_pop
```

```python
plt.figure(figsize=(10,6))
total_pop.head(10).plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('Top 10 Countries with Highest Population(2022)')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x=total_pop.head(10).index,y=total_pop.head(10).values,palette='coolwarm')
plt.title('Top 10 Countries with Highest Population(2022)')
plt.xlabel('Country')
plt.ylabel('Total Population(2022)')
plt.show()
```

#Bottom 10 countries with respect to total population

```python
total_pop1=data.groupby('Country/Territory')['2020 Population'].sum().sort_values(ascending=True)
total_pop1
```

```python
plt.figure(figsize=(10,6))
total_pop1.head(10).plot(kind='bar',color='skyblue',edgecolor='black')
plt.title('Total Population by Country(2022)')
plt.xlabel('Country')
plt.ylabel('Total Population')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x=total_pop1.head(10).index,y=total_pop1.head(10).values,palette='coolwarm')
plt.title('Total Population by Country(2022)')
plt.xlabel('Country')
plt.ylabel('Total Population(2022)')
plt.show()
```


This visualization demonstrates both categorical and continuous data distributions effectively.

A histogram helps to visualize the frequency of countries within specific population ranges.

Bar charts are ideal for comparing values across different categories.
